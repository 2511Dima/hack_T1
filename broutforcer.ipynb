```python
!pip3 install scikit-learn
!pip3 install numpy
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans


df = pd.read_csv('clinic-data.csv', delimiter=',')
df.head()

x = list()
y = list()
for x1 in df['x']:
    x.append(x1)
for y1 in df['y']:
    y.append(y1)

data = df[['x','y']].values
```

```python
res = list()
for i in range(1, 4000):
    rand = i
    kmeans = KMeans(n_clusters=250, random_state=rand, n_init=20)
    kmeans.fit(data)
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    diff = data[:, None, :] - centroids[None, :, :] #broadcasting
    dist = np.linalg.norm(diff, axis=2) #евклидово растояние
    nearest_building_idx = dist.argmin(axis=1) 
    a = 0.05
    b = 0.02

    cost = a * dist + b * dist**2
    min_cost_per_client = cost.min(axis=1)
    ttt = min_cost_per_client.sum()
    print('Выполнено:',i, ttt)
    res.append([float(ttt),rand])

find_min = list()
find_min_id = list()
for i in res:
    find_min_id.append(i[1])
    find_min.append(i[0])
id_min = find_min.index(min(find_min))
print('\nРЕЗУЛЬТАТ, минимальтый ТТТ -', min(find_min), find_min_id[id_min])
```
